```
import tensorflow as tf
```

```
#Tensorflow works as a graph computational model
#Building a graph in Tensorflow
```

```
a = tf.constant([2])
b = tf.constant([3])
c = tf.add(a, b)
```

```
#Tensorflow needs to initialize a Session to run the code. i.e they are a context for creating a graph inside tensorflow

session = tf.Session()
```

```
#run session to get result from previously defined c operation
result = session.run(c)
print(result)
```

```
#close the session to release resources
session.close()
```

```
#use a *with* block to close session automatically

with tf.Session() as session:
  result = session.run(c)
  print(result)
```
